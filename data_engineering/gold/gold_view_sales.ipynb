# Gold Layer - Analytics View: `vw_sales`

This notebook creates a business-facing analytics view by joining Gold sales with customer and product dimensions.

## Tools and Operations
- **Spark SQL** view definition
- Standardized semantic layer for BI/dashboard queries

## Output
- `workspace.gold.vw_sales`


## 1) Create or Replace Gold View


```python
%sql
CREATE OR REPLACE VIEW workspace.gold.vw_sales AS
SELECT 
  s.*,
  p.product_name,
  p.product_line,
  p.category_name,
  p.subcategory_name,
  c.gender,
  c.marital_status,
  c.country
FROM gold.sales s
LEFT JOIN gold.customers c USING (customer_key) 
LEFT JOIN gold.products p USING (product_key)
```
